# Understanding CNN Architecture: How Each Layer Works

Let's walk through each layer of the network. Each step plays a crucial role in transforming a raw input image into a final classification prediction.

## Input Layer
The input layer represents the input image into the CNN. Because we use RGB images as input, the input layer has three channels, corresponding to the red, green, and blue color channels. Detailed pixel values and colors can be displayed to understand the raw data feeding into the network.

## Convolutional Layers
The convolutional layers are the foundation of a CNN. They contain the learned kernels (weights) that extract features distinguishing different images from one another—which is exactly what we want for classification! 

The convolutional neuron performs an elementwise dot product with a unique kernel and the output of the previous layer’s corresponding neuron. This will yield as many intermediate results as there are unique kernels. The convolutional neuron is the result of all of the intermediate results summed together with the learned bias.

For example, let’s look at the first convolutional layer in the Tiny VGG architecture above. Notice that there are 10 neurons in this layer, but only 3 neurons in the previous layer. In the Tiny VGG architecture, convolutional layers are fully-connected, meaning each neuron is connected to every other neuron in the previous layer. Focusing on the output of the topmost convolutional neuron from the first convolutional layer, we see that there are 3 unique kernels when we hover over the activation map.

![Working of CNN](https://poloclub.github.io/cnn-explainer/assets/figures/convlayer_overview_demo.gif)

The size of these kernels is a hyper-parameter specified by the designers of the network architecture. In order to produce the output of the convolutional neuron (activation map), we must perform an elementwise dot product with the output of the previous layer and the unique kernel learned by the network. In TinyVGG, the dot product operation uses a stride of 1, which means that the kernel is shifted over 1 pixel per dot product, but this is a hyperparameter that the network architecture designer can adjust to better fit their dataset. We must do this for all 3 kernels, which will yield 3 intermediate results.

![CNN working](https://poloclub.github.io/cnn-explainer/assets/figures/convlayer_detailedview_demo.gif)

Then, an elementwise sum is performed containing all 3 intermediate results along with the bias the network has learned. After this, the resulting 2-dimensional tensor will be the activation map viewable on the interface above for the topmost neuron in the first convolutional layer. This same operation must be applied to produce each neuron’s activation map.

With some simple math, we are able to deduce that there are 3 x 10 = 30 unique kernels, each of size 3x3, applied in the first convolutional layer. The connectivity between the convolutional layer and the previous layer is a design decision when building a network architecture, which will affect the number of kernels per convolutional layer.

## Understanding Hyperparameters

### Input Size
The dimensions of the input feature map (Height, Width).

### Padding
Padding is often necessary when the kernel extends beyond the edges of the activation map. Symmetrically adding zeros around the edges (zero-padding) conserves data at the borders of activation maps, leading to better performance and preserving the input's spatial size so you can build deeper networks.

### Kernel Size (Filter Size)
The dimensions of the sliding window over the input. Small kernel sizes (like 3x3) extract highly local features, leading to a smaller reduction in layer dimensions, which allows designers to stack more layers. Large kernels extract larger, global features but reduce layer dimensions much faster.

### Stride
Indicates how many pixels the kernel shifts at a time. A smaller stride (like 1) extracts more detailed data but results in larger output layers. A larger stride decreases feature extraction density and shrinks output layer dimensions rapidly. 

## Activation Functions

### ReLU (Rectified Linear Unit)
High-performing CNNs achieve high accuracy because of their non-linearity. ReLU applies much-needed non-linearity to the model, which is necessary to produce non-linear decision boundaries so the output cannot be written as a simple linear combination of inputs. Without non-linear activation functions, a deep CNN would collapse into a single equivalent linear layer.

ReLU is preferred over Sigmoid because networks using ReLU are significantly faster to train. It is a simple elementwise operation: if a value is positive, it passes through unchanged; if it is negative, it becomes exactly 0.

### Softmax
A softmax operation serves a key purpose: making sure the output predictions of the network sum to 1, allowing us to interpret them as probabilities. 

Before passing through the softmax function, the raw outputs (called logits) are unscaled scalar values. Softmax rescales these logits between 0 and 1. 

Unlike standard normalization, softmax acts as a differentiable approximation of the "argmax" function. It exponentiates the logits, which heavily weighs the maximum value relative to others, ensuring the correct answer stands out with a strong signal during backpropagation.

## Pooling Layers
Pooling layers gradually decrease the spatial extent (width and height) of the feature maps, reducing the number of parameters and computation in the network to avoid overfitting.

The Max-Pooling operation slides a kernel with a specified stride over the input while only selecting the largest value at each step. In the Tiny VGG architecture, the pooling layers use a 2x2 kernel and a stride of 2, which discards 75% of the activations to improve computational efficiency.

## Flatten Layer
This layer converts a three-dimensional layer in the network (Height x Width x Channels) into a one-dimensional vector to fit the input of a fully-connected layer for classification. 

For example, a 5x5x2 tensor is flattened into a vector of size 50. While the convolutional layers extract the features, the fully-connected layers and the softmax function classify these features, which requires a 1-dimensional input.
