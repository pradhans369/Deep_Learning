# **RNN Forward Propagation: 3 example sentences (i like cats, cats look good, some cats are white)**


**REMEMBER** : 
- In ML we use vectorization of words to convert words into ints. But in DL we also use **Embeddings** to convert words into numbers (float).


## 1. Parameters & Initialization
- **Weight Matrices (Fixed):**
  - $W_{xh} = \begin{bmatrix} 0.5 & 0.0 \\ 0.0 & 0.5 \end{bmatrix}$ (Input Weights)
  - $W_{hh} = \begin{bmatrix} 0.1 & 0.0 \\ 0.0 & 0.1 \end{bmatrix}$ (Recurrent Weights)
- **Initial State:** $h_0 = [0, 0]$
- **Activation Function:** $\tanh$


## 2. Sentence 1: "i like cats"
**Word Vectors:** - "i" = [0.1, 0.2]
- "like" = [0.3, 0.1]
- "cats" = [0.5, 0.4]

### Mathematical Steps:
1. **Step 1 ("i"):**
   $h_1 = \tanh(W_{xh} \cdot [0.1, 0.2] + W_{hh} \cdot [0, 0]) = \tanh([0.05, 0.1]) \approx [0.05, 0.10]$

2. **Step 2 ("like"):**
   $h_2 = \tanh(W_{xh} \cdot [0.3, 0.1] + W_{hh} \cdot [0.05, 0.1]) = \tanh([0.15+0.005, 0.05+0.01]) \approx [0.15, 0.06]$

3. **Step 3 ("cats"):**
   $h_3 = \tanh(W_{xh} \cdot [0.5, 0.4] + W_{hh} \cdot [0.15, 0.06]) = \tanh([0.25+0.015, 0.2+0.006]) \approx [0.26, 0.20]$


## 3. Sentence 2: "cats look good"
**Word Vectors:**
- "cats" = [0.5, 0.4]
- "look" = [0.2, 0.3]
- "good" = [0.4, 0.5]

### Mathematical Steps:
1. **Step 1 ("cats"):**
   $h_1 = \tanh(W_{xh} \cdot [0.5, 0.4] + [0, 0]) = \tanh([0.25, 0.2]) \approx [0.24, 0.20]$

2. **Step 2 ("look"):**
   $h_2 = \tanh(W_{xh} \cdot [0.2, 0.3] + W_{hh} \cdot [0.24, 0.20]) = \tanh([0.1+0.024, 0.15+0.02]) \approx [0.12, 0.16]$

3. **Step 3 ("good"):**
   $h_3 = \tanh(W_{xh} \cdot [0.4, 0.5] + W_{hh} \cdot [0.12, 0.16]) = \tanh([0.2+0.012, 0.25+0.016]) \approx [0.21, 0.26]$

# **How output at t-1 is used at the different neuron ???**


$$h_t = \tanh(W_{xh} x_t + W_{hh} h_{t-1} + b)$$

- **$h_t$**: The hidden state vector at time $t$ (The "Memory").
- **$x_t$**: The input vector at time $t$.
- **$W_{xh}$**: Weight matrix mapping inputs to hidden states.
- **$W_{hh}$**: Weight matrix mapping previous memory to current memory (The "Recurrent Connection").
- **$b$**: Bias vector.
- **$\tanh$**: Non-linear activation squashing values between -1 and 1.

#### Neuron-Level Interaction
In an RNN layer with $N$ neurons:

1. **The Shared Memory**: $h_{t-1}$ is not a private logbook; it is a vector of $N$ values.
2. **Matrix Multiplication**: The term $W_{hh} h_{t-1}$ is a matrix-vector multiplication. Each neuron $i$ calculates its contribution from the *entire* previous layer by taking a dot product of its specific row in $W_{hh}$ and the entire vector $h_{t-1}$.
3. **Interconnectivity**: Because every neuron's output at $t-1$ is part of the $h_{t-1}$ vector, every neuron's past state influences every neuron's current state.

**BASCIALLY THE OUTPUT AT T-1 HOLD THE OUTPUT OF ALL THE NEURONS AT T-1 STEP OF THE SAME LAYER.**

## 3. Full Lifecycle Summary (Step-by-Step)

| Step | Process | Action |
| :--- | :--- | :--- |
| **0** | **Initialization** | Set $h_0 = \vec{0}$ (The blank slate). |
| **1..T** | **Recurrent Loop** | For each step $t$: Multiply $x_t$ by $W_{xh}$, multiply $h_{t-1}$ by $W_{hh}$, sum them with $b$, and apply $\tanh$. |
| **T** | **Output Projection** | After reading the final input, compute $y_T = \text{softmax}(W_{hy} h_T + b_y)$. |

## 4. Example Mathematical Progression
*Using $W_{xh}=0.5$, $W_{hh}=0.1$, $h_0=[0,0]$ for simplicity.*

### Processing "i like cats"
1. **Step 1 ("i" = [0.1, 0.2])**:
   $h_1 = \tanh(0.5 \cdot [0.1, 0.2] + 0.1 \cdot [0, 0]) \approx [0.05, 0.10]$

2. **Step 2 ("like" = [0.3, 0.1])**:
   $h_2 = \tanh(0.5 \cdot [0.3, 0.1] + 0.1 \cdot [0.05, 0.10]) \approx [0.15, 0.06]$

3. **Step 3 ("cats" = [0.5, 0.4])**:
   $h_3 = \tanh(0.5 \cdot [0.5, 0.4] + 0.1 \cdot [0.15, 0.06]) \approx [0.26, 0.20]$

**Result**: The final memory state $h_3$ is $[0.26, 0.20]$, which represents the mathematical summary of the sequence "i like cats".

## **Why do we call it 'Recurrent'**


**[COZ WE ARE USING THE OUPUT OF THE A NEURON IN OTHER NEURONS IN THE SAME LAYER AGAIN AND AGAIN]**

We call it a Recurrent Neural Network (RNN) because of the way the model handles data over time: it uses the same operation repeatedly (recurrently) on a sequence of inputs, while maintaining a "memory" of what it saw previously.

In a standard "feed-forward" neural network, information moves in only one direction: from the input to the output. Once the calculation is done, the network "forgets" everything.

In an RNN, the output of the hidden layer is fed back into itself as an input for the next time step. Mathematically, this is expressed as:

$$h_t = f(h_{t-1}, x_t)$$

Because the model uses the previous state ($h_{t-1}$) to calculate the new state ($h_t$), it creates a feedback loop. This loop is what we mean by "recurrent"—it is a process that feeds itself.

The term "recurrent" is used because:
- Time-Recurrence: The network performs the same functional transformation at every time step $t$.
- Shared Parameters: Because it is the same operation, the weight matrices ($W_{xh}, W_{hh}$) are reused (shared) across every single step in the sequence. It is not creating new neurons for every word; it is recycling the same neurons over and over.
