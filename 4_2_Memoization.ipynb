# Memoization:



Memoization is an optimization technique used primarily to speed up computer programs by storing the results of expensive function calls and returning the cached result when the same inputs occur again.

Memoization is a specific form of **caching** used to speed up computer programs by storing the results of expensive function calls and returning the cached result when the same inputs occur again.

## 1. The Core Concept
When a function is "expensive" (computationally intensive or time-consuming), recalculating it for the same input is redundant. Memoization solves this by:

1.  **Checking:** Before executing the function, the program checks a cache (usually a dictionary or hash map) to see if the result for these specific inputs exists.
2.  **Returning:** If found, it returns the cached result instantly.
3.  **Calculating & Storing:** If not found, it performs the calculation, stores the result in the cache, and then returns it.



## 2. When to Use It
Memoization is highly effective when:
* **Determinism:** The function is pure (the same input always produces the same output).
* **Repetition:** The function is called multiple times with the same parameters.
* **Resource Trade-off:** You are willing to trade **Memory (RAM)** for **Speed (CPU time)**.

## 3. Example: Fibonacci Sequence
Without memoization, calculating Fibonacci numbers using recursion is extremely inefficient because it re-calculates the same values multiple times.

* **Without Memoization:** $O(2^n)$ time complexity (exponential).
* **With Memoization:** $O(n)$ time complexity (linear).



## 4. Implementation Example (Python)

Python provides a built-in decorator to handle memoization automatically using `functools`.

```python
from functools import lru_cache

# The @lru_cache decorator handles the memoization logic for you
@lru_cache(maxsize=None)
def expensive_fibonacci(n):
    if n < 2:
        return n
    return expensive_fibonacci(n-1) + expensive_fibonacci(n-2)

# First call calculates the value
print(expensive_fibonacci(30))

# Subsequent calls for 30 are returned instantly from the cache
print(expensive_fibonacci(30))
