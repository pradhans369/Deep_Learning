Types of RNN:

1. Many to One
    - Sentiment analysis, 
    - Rating Prediction based on movie reviews
    - Spam detection
    - Activity recognition from video
2. One to Many
    - Image captioning
    - Music generation
    - Image to text formatting (getting texts from img)
3. Many to Many (also called encoder decoder model)
    - Part of speech tagging (tagging grammar in the text)
    - Stock price prediction


Many to many is also of two types:
- Same length inuts and outputs
- Different length inputs and outputs
