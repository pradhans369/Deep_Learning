# Improving performance of a neural network

1. Fine tuning hyperparameters
    - setting the learning rate
        - use **learning rate scheduler**. If you are using large batch size, at the initial epoch keep the learning rate low, but the later epochs keep increasing the learning rate, this trains the model faster
        - 
    - setting number of hidden layers
        - instead of taking lot of leurons with less hidden layers, just take the lot of hidden layers with less number of neurons
    - setting number of neurons per layer
        - there should be sufficient number of neurons
        - if there is 8 input features then at the first hidden layer the number of neurons should be mroe than 8, so they can capture every hidden pattern in the data
    - setting the optimizer
        - Adam
    - setting the regularizers
        - l1 / l2
    - setting the batch size
    - setting the activation function
    - setting the epochs
        - for higher epochs use early stopping
    - use optuna for finding best parameters


2. Solving the problems :
    - vanishing gradient / exploding gradient
        - using weight initialization
        - changing activation function
        - using batch normalization
        - usign gradient clipping for exploding gradient
    - lack of data
        - using transfer learning (using the model made by someone else)
        - using unsupervised pre-training
    - slow training
        - using Adam, momentum, adagrad, RMSprop
        - using learning rate scheduler
    - overfitting
        - using l1 / l2
        - using dropout
    - imbalane dataset
        - using SMOTE
    - outliers
        - capping etc
