# **Transfer Learning**

Transfet Learning is a method where you use a pre-trained model. A model which was trained on a different dataset **(Here, you take the only conv layers of the pretrained model)** and you apply it on another dataset of similar kind to make the prediction.

There are type of Transfer Learning:
1. Feature Extraction
    - Here you freeze the entire conv layers of the pretrained model, and only train the dense layers.
    - YOU FOLLOW THIS APPROACH WHEN THE LABELS YOU WANT TO PREDICT ARE THE SAME LABELS ON WHICH THE PRE-TRAINED MODEL WAS TRAINED ON.
2. Fine Tuning the model
    - Here you donot freeze the entire conv layers of the pretrained model, but rather unfreeze only last few layers , and only train the dense and last few unfreezed conv layers.
    - YOU FOLLOW THIS APPROACH WHEN THE LABELS YOU WANT TO PREDICT ARE NOT SAME LABELS ON WHICH THE PRE-TRAINED MODEL WAS TRAINED ON. **(predicting phones on a model which was trained on animals)**

## **Why We Freeze Convolutional Layers in Transfer Learning ???**

1. **Feature Reuse (General vs. Specific Features)**
   * The early convolutional layers of a pre-trained model detect generic features like edges, curves, textures, and color gradients. 
   * These features are universal and apply to almost any image dataset (whether it is ImageNet, CIFAR-10, or custom medical scans). There is no need to spend time relearning them.

2. **Preventing Feature Destruction (Catastrophic Forgetting)**
   * When you attach a new classification head (e.g., a `Dense(100)` layer), its weights are initialized completely at random.
   * If you train the entire model immediately, the massive gradients (errors) coming from this random layer will flow backward through the network, overwriting and destroying the valuable pre-trained features in the backbone. Freezing protects those weights.

3. **Speed and Efficiency**
   * Frozen weights do not require gradients to be calculated or stored during backpropagation. 
   * This drastically reduces the computational workload, allowing epochs to run significantly faster (e.g., dropping from 71 seconds to 15 seconds on a GPU).

4. **Prevents Overfitting on Small Datasets**
   * Deep convolutional backbones have millions of parameters. 
   * If you try to train all of them on a small dataset, the model will quickly memorize the training set. Freezing the convolutions restricts the trainable parameters to only the final classifier layer, acting as a strong regularizer.
