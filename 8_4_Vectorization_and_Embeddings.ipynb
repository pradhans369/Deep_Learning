```python
import numpy as np

# creating dataset
docs = ["I love deep learning",
        "NLP is awesome",
        "Recurrent neural networks have memory"]
```

```python
# applying tokenization on words (each word will be converted into a different number)
from keras.layers import TextVectorization

vectorizer = TextVectorization(output_mode='int')
vectorizer.adapt(docs)

vectorized_text = vectorizer(docs)
```

```python
print(vectorized_text)
```

### Type 1

```python
temp = vectorizer.get_vocabulary()
for i, word in enumerate(temp):
    print(f"{word} : {i}")
```

### Type 2

```python
vectorizer = TextVectorization(output_mode='count')
vectorizer.adapt(docs)

vectorized_text2 = vectorizer(docs)
```

```python
print(vectorized_text2)
```

```python
for i, word in enumerate(vectorizer.get_vocabulary()):
    print(f"{word} : {i}")
```

### Type 3

```python
vectorizer = TextVectorization(output_mode="tf_idf")
vectorizer.adapt(docs)

vectorized_text3 = vectorizer(docs)
```

```python
print(vectorized_text3.shape)
print()
print(vectorized_text3.numpy())
```

**Using tfidf will create sparsity in the dataset, i mean if there is only 2 words in a sentence but still there will be 13 features, rest 11 features will be 0. same issue with count. And using int, is not good coz, model will prioritize the word with highest number. To prevent these Embedding is used.**


**EMBEDDING : A word embedding is a dense, low-dimensional vector representation of text where words with similar meanings or semantic relationships are mapped to neighboring points in a continuous vector space. In Embedding all the values are non zeros, not like tfidf or count where some of the values are zeros.**

### Type 4 - Embeddings

```python
from keras import layers
```

```python
vectorizer = TextVectorization(output_mode='int')
vectorizer.adapt(docs)
vectorized_text = vectorizer(docs)

length = len(vectorizer.get_vocabulary())

embedding = layers.Embedding(input_dim=length, output_dim=4)

vectorized_text = embedding(vectorized_text)
```

```python
print(vectorized_text.shape)
print()
print(vectorized_text)
```
