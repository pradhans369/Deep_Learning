# Backpropagation

Backpropagation (Backward Propagation of Errors) is the fundamental algorithm used to train neural networks. It calculates the **gradient** of the loss function with respect to every weight in the network by applying the **Chain Rule** from calculus.

## 1. The Big Picture
Training a neural network happens in three repeating steps:
1.  **Forward Pass:** Data flows through the network to generate a prediction ($\hat{y}$).
2.  **Loss Calculation:** The prediction is compared to the truth ($y$) using a Loss Function (like Binary Cross-Entropy).
3.  **Backward Pass (Backpropagation):** The "blame" for the error is sent backward through the layers to update the weights.



## 2. Step 1: The Forward Pass
In a single neuron, we calculate the weighted sum of inputs plus a bias ($z$) and pass it through an activation function ($\sigma$).

### The Equations:
1.  **Linear Combination:** $$z = \sum (w_i \cdot x_i) + b$$
2.  **Activation (e.g., Sigmoid):** $$\hat{y} = \sigma(z) = \frac{1}{1 + e^{-z}}$$



## 3. Step 2: Calculating Loss
For classification, we use **Binary Cross-Entropy (BCE)**. It penalizes the model heavily if it is confident and wrong.

$$L = -[y \log(\hat{y}) + (1 - y) \log(1 - \hat{y})]$$

## 4. Step 3: The Backward Pass (The Math)
The goal is to find how a change in weight ($w$) affects the Loss ($L$). We use the **Chain Rule** to break this down into three connected links.



### The Chain Rule Formula:
$$\frac{\partial L}{\partial w} = \frac{\partial L}{\partial \hat{y}} \times \frac{\partial \hat{y}}{\partial z} \times \frac{\partial z}{\partial w}$$

1.  **Link 1 (Loss Gradient):** How much does the Loss change if the prediction changes?
    $$\frac{\partial L}{\partial \hat{y}} = \frac{\hat{y} - y}{\hat{y}(1 - \hat{y})}$$
2.  **Link 2 (Activation Gradient):** How much does the prediction change if $z$ changes?
    $$\frac{\partial \hat{y}}{\partial z} = \sigma(z)(1 - \sigma(z))$$
3.  **Link 3 (Linear Gradient):** How much does $z$ change if the weight changes?
    $$\frac{\partial z}{\partial w} = x$$

### The "Magic" Simplification:
When you multiply these together for Sigmoid + BCE, the denominators cancel out perfectly, leaving a very clean gradient:
$$\frac{\partial L}{\partial w} = (\hat{y} - y) \cdot x$$

## 5. Step 4: Weight Update
Once we have the gradient, we update the weight using the **Learning Rate ($\alpha$)**. We move in the *opposite* direction of the gradient to find the minimum of the "Loss Valley."

$$w_{new} = w_{old} - \alpha \cdot \frac{\partial L}{\partial w}$$

# The Complete Workflow of Backpropagation

Backpropagation is an iterative process used to train neural networks by calculating the gradient of the loss function with respect to every weight in the network.

---

## 1. Initialization
* **Define Loss Function:** Choose based on your objective.
    * *Classification:* Binary Cross-Entropy (BCE), Categorical Cross-Entropy (CCE).
    * *Regression:* Mean Squared Error (MSE), Mean Absolute Error (MAE).
* **Set Hyperparameters:** Define the number of epochs and the Learning Rate ($\alpha$).

---

## 2. The Training Loop (Epochs)
The model iterates through the training data for a fixed number of epochs. Within each epoch:

### A. Mini-Batch Selection
* Instead of picking 1 sample (SGD), we select a **Mini-Batch** (e.g., 32 or 64 samples). This stabilizes learning and optimizes GPU usage.

### B. The Forward Pass
1.  **Linear Transformation:** For each layer, compute $z = (w \cdot x) + b$.
2.  **Activation:** Apply non-linear functions (ReLU, Sigmoid, etc.) to produce the layer output.
3.  **Propagation:** Pass the output to the next layer until the final layer generates $\hat{y}$.
4.  **Caution:** Use `Softmax/Sigmoid` for classification and `ReLU/Linear` for regression in the final layer.

### C. Error Calculation
* Compare the prediction ($\hat{y}$) against the ground truth ($y$) using the chosen loss function.
    * $L = \text{Loss}(\hat{y}, y)$

---

## 3. The Backward Pass (Backpropagation)
This is where the "learning" happens. We use the **Chain Rule** to propagate the error backward.

1.  **Gradient Calculation:** Starting from the final layer, calculate the partial derivative of the Loss with respect to each weight:
    $$\frac{\partial L}{\partial w} = \frac{\partial L}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial z} \cdot \frac{\partial z}{\partial w}$$
2.  **Weight Adjustment:** Update the weights using the Gradient Descent rule:
    $$w_{new} = w_{old} - (\alpha \cdot \frac{\partial L}{\partial w})$$
    * *Note:* We do not aim for an error of 0, as that causes **overfitting**. We aim for the global minimum on the validation set.

---

## 4. Evaluation and Completion
* **Epoch Completion:** Once all samples in the batch/dataset have been processed and weights adjusted, one epoch is complete.
* **Validation:** Calculate the loss/accuracy on a separate test set to ensure the model is generalizing, not just memorizing.
* **Inference:** Once the desired loss is achieved or the training loop finishes, the weights are "frozen." The model is now ready to predict $\hat{y}$ for unseen data.
