# How to name the perceptrons and weights

- `m` = number of rows
- `n` = total number of features
- `x_i1` = first value of the `ith` row
- `x_i2` = second value of the `ith` row
- `o_11` = output of the first node in first layer
- `o_13` = output of the third node in first layer
- `w^1 21` = wight entering into first layer, coming from second node and entering into first node
- `w^2 13` = wight entering into second layer, coming from first node of previous layer and entering into third node

1.  At input layer each feature send the inputs with weights to different neuron at first layer.
2.  At the first layer each neuron holds the formula for neuron and a threshold function. And hence each neuron produces a `z` output.
3.  Then at second layer each neuron receives different output (`z`) and weights from different neurons from layer 1. I just multiplies the output 1 with its weights, and multiplies output 2 with its respective weights and does the same with all the output received from first layer.
    - And after multiplying the output with their respective weights it just makes a sum of those values (THIS IS CALLED `DOT PRODUCT`) and adds a **bias** at the end (`b` in `(mx + b)`).
    - And then a threshold function is used to get `z`. And that is how each neuron in layer two produces `z`
4.  After layer two, layer three does the same as layer two received the different `z` with their respective weights, sums it all up and adds a bias at the end, and produces it own `z`. 
5. And this is how a final ans is produced. By doing this, the DL multi layer model is able to capture hidden patterns in the data, which were being missing by the single neuron model. Here if one pattern is missed by one neuron then it will be cought by another neuron.
5.  **MORE THE NEURON AND LAYERS MORE WILL BE THE ACCURACY, BUT MORE WILL BE OVERFITTING, SLOW, AND COMPUTATIONAL COST.** 

####  **The entire structure from input layer to the output layer is one multi layer perceptron**
