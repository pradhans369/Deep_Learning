# Vanishing Gradient Problem


The **Vanishing Gradient Problem** is one of the most famous roadblocks in Deep Learning. It is the main reason why, for a long time, researchers couldn't build Neural Networks that were very "deep" (networks with many hidden layers).


for eg, 0.1 x 0.2 x 0.5 = 0.01 (the ans is smaller than the numbers itself, while the ans was supposed to be higher than the numbers as its being multiplied)<br>
This happens when `sigmoid function` is used in the hidden layers.


Here is the easiest way to understand it, broken down step-by-step:

### 1. How Learning Usually Works (The Chain Rule)
When a neural network makes a mistake, it uses **Backpropagation** to send an "error signal" backward from the output layer all the way to the first input layer. 
To pass this signal backward through the layers, the network uses calculus (the Chain Rule), which means it **multiplies** the gradients (slopes) of each layer together.

### 2. The Cause: Multiplying Tiny Fractions
Older activation functions, like the **Sigmoid** function, squash numbers between `0` and `1`. 
Because of how the math works, the maximum possible gradient (derivative) of a Sigmoid function is **`0.25`**. 

Imagine you have a deep network with 10 layers. To find out how much to update the weights in the very first layer, backpropagation has to multiply those gradients together 10 times:
`0.25 × 0.20 × 0.15 × 0.25 × 0.10...`

When you multiply small fractions together, the number shrinks incredibly fast. By the time that error signal reaches the first layer, the gradient is `0.000000001`. **It has vanished.**

### 3. The Result: The Network Gets "Stuck"
The formula to update a neural network's weights is:
`New Weight = Old Weight - (Learning Rate × Gradient)`

If the Gradient has shrunk to `0.000000001`, the network calculates:
`New Weight = Old Weight - (0)`

The weights in the early layers don't change. **The first layers completely stop learning.** If the first layers can't learn how to detect basic patterns (like edges or shapes), the later layers have garbage data to work with, and the entire AI fails to train.



### How do we fix it today?
This problem paralyzed AI development until a few massive breakthroughs occurred:

1. **Changing the Activation Function (ReLU):** Instead of Sigmoid, we now use **ReLU** for hidden layers. The derivative of ReLU is exactly `1.0` (for positive numbers). Multiplying `1.0 × 1.0 × 1.0` never vanishes, allowing the error signal to travel backward through hundreds of layers safely.
2. **ResNets (Skip Connections):** Creating "express lanes" in the network so the gradient can skip layers and jump straight to the beginning without being mathematically squashed.
3. **Better Weight Initialization:** Using techniques like **He Initialization** to ensure starting weights don't accidentally shrink the gradients from the very first epoch.
