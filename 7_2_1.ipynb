```python
import keras
from keras import models, layers, Sequential
from keras.layers import Dense, Conv2D, Flatten
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
model = models.Sequential([
    layers.Conv2D(32, kernel_size=(3,3), padding='valid', strides=(2,2), activation='relu', input_shape=(28,28,1)),
    layers.Conv2D(32, kernel_size=(3,3), padding='valid', strides=(2,2), activation='relu'),
    layers.Conv2D(32, kernel_size=(3,3), padding='valid', strides=(2,2), activation='relu'),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
model.summary()
```

```python
model = models.Sequential([
    layers.Conv2D(32, kernel_size=(3,3), padding='same', activation='relu', strides=(2,2), input_shape=(28,28,1)),
    layers.Conv2D(32, kernel_size=(3,3), padding='same', strides=(2,2), activation='relu'),
    layers.Conv2D(32, kernel_size=(3,3), padding='same', strides=(2,2), activation='relu'),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```
