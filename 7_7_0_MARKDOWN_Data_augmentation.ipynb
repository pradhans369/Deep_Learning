# **Data Augmentation**

Keras provides several layers to apply different types of transformations to images. They can be divided into three main categories:

### 1. Geometric Transformations (Changing Shape, Position, & Angle)
* **RandomFlip**: Flips the image horizontally, vertically, or both.
  * *Code*: `layers.RandomFlip("horizontal_and_vertical")`
* **RandomRotation**: Rotates the image by a random fraction of 360 degrees.
  * *Code*: `layers.RandomRotation(factor=0.2)` (Rotates up to 72 degrees)
* **RandomZoom**: Zooms in or out of the image randomly.
  * *Code*: `layers.RandomZoom(height_factor=0.2, width_factor=0.2)`
* **RandomTranslation**: Shuts or shifts the image horizontally or vertically (moving the object off-center).
  * *Code*: `layers.RandomTranslation(height_factor=0.1, width_factor=0.1)`

### 2. Color and Contrast Modifications (Changing Lighting & Tone)
* **RandomContrast**: Randomly increases or decreases the contrast (makes darks darker and brights brighter).
  * *Code*: `layers.RandomContrast(factor=0.2)`
* **RandomBrightness**: Randomly adjusts the brightness of the image (simulates different lighting conditions).
  * *Code*: `layers.RandomBrightness(factor=0.2)`

### 3. Cropping and Resizing (Simulates Camera framing)
* **RandomCrop**: Randomly crops a smaller section of the image and resizes it back to the input shape.
  * *Code*: `layers.RandomCrop(height=200, width=200)`

## **Data Augmentation in Keras**

Data augmentation is a technique used to artificially expand the size of a training dataset by creating modified versions of images in the dataset.

Here is the complete code to define and use Data Augmentation in your CNN model:

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Sequential

# 1. Define the data augmentation pipeline as a Keras Sequential block
# (Keras automatically disables these transformations during validation, testing, and predictions)
data_augmentation = Sequential([
    # Randomly flip images horizontally (mirroring)
    layers.RandomFlip("horizontal", input_shape=(256, 256, 3)),
    
    # Randomly rotate images by up to 15% (about 54 degrees)
    layers.RandomRotation(0.15),
    
    # Randomly zoom in or out by up to 15%
    layers.RandomZoom(0.15)
])

# 2. Insert the data_augmentation block as the first step of your CNN model
model = Sequential([
    # Step 1: Augment the training images
    data_augmentation,

    # Step 2: Rescale pixel values from [0, 255] to [0.0, 1.0]
    layers.Rescaling(1./255),

    # Step 3: Convolutional and Pooling layers
    layers.Conv2D(32, kernel_size=(3, 3), padding='valid', activation='relu'),
    layers.MaxPooling2D(pool_size=(2, 2), strides=2),

    layers.Conv2D(64, kernel_size=(3, 3), padding='valid', activation='relu'),
    layers.MaxPooling2D((2, 2), strides=2),
    
    layers.Conv2D(128, kernel_size=(3, 3), padding='valid', activation='relu'),
    layers.MaxPooling2D((2, 2), strides=2),
    
    # Step 4: Flattening and Fully Connected Layers with Dropout
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),  # Prevents overfitting by shutting down 30% of neurons
    
    layers.Dense(64, activation='relu'),
    
    # Step 5: Output Layer (Cats = 0, Dogs = 1)
    layers.Dense(1, activation='sigmoid')
])
