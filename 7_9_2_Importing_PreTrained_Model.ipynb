# Importing Pre-Trained Models

Since keras has updated its database for pre-trained models;
1. Classical way of importing pre-trained models (traditional type)
2. **Keras Hub** way of importing pre-trained models (new type)

# Type 1 : Classical way of importing pre-trained model

```python
from keras import layers, Sequential
from keras.applications.resnet50 import ResNet50
```

```python
# importing ResNet model
base_model = ResNet50(
    weights='imagenet',
    include_top=False,
    input_shape=(224, 224, 3)
)

# Freezing the base model weights so that they donot change during backpropagation
base_model.trainable=False
```

```python
# creating a new model, where after adding the base_model we will add the custom model
model = Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(1, activation='sigmoid')
])
```

```python
# compiling the model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```
