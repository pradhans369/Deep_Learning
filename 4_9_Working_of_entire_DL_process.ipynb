# Mathematical Derivation of Dataset Derivatives (Gradient Descent Example)

Here is the exact, step-by-step calculus for training a simple machine learning model on a housing dataset using Mean Squared Error (MSE) and Gradient Descent.

---

### 1. Defining the Mathematical Setup

* **Input ($x_i$):** The size of house $i$ (in Sq Ft).
* **Actual Output ($y_i$):** The actual price of house $i$ (in Dollars).
* **Model Equation ($\hat{y}_i$):** Our prediction. 
  $$\hat{y}_i = w \cdot x_i$$
  *(Where $w$ is our single parameter: price per square foot. We assume bias $b = 0$ for simplicity).*

---

### 2. The Loss Function: Mean Squared Error (MSE)

To measure how wrong our model is across all $N$ houses, we use the **Mean Squared Error (MSE)** loss function:

$$L(w) = \frac{1}{N} \sum_{i=1}^{N} (\hat{y}_i - y_i)^2$$

Substitute $\hat{y}_i = w \cdot x_i$ into the equation:

$$L(w) = \frac{1}{N} \sum_{i=1}^{N} (w \cdot x_i - y_i)^2$$

Since we have $N = 3$ houses in our dataset, the formula expands to:

$$L(w) = \frac{1}{3} \left[ (w \cdot x_1 - y_1)^2 + (w \cdot x_2 - y_2)^2 + (w \cdot x_3 - y_3)^2 \right]$$

---

### 3. Calculating the Loss for $w = 100$

Let's plug in our starting weight **$w = 100$** and our dataset values:
* **House 1:** $x_1 = 1,000$, $y_1 = 150,000$
* **House 2:** $x_2 = 1,500$, $y_2 = 220,000$
* **House 3:** $x_3 = 2,000$, $y_3 = 290,000$

$$L(100) = \frac{1}{3} \left[ (100 \cdot 1000 - 150,000)^2 + (100 \cdot 1500 - 220,000)^2 + (100 \cdot 2000 - 290,000)^2 \right]$$

$$L(100) = \frac{1}{3} \left[ (-50,000)^2 + (-70,000)^2 + (-90,000)^2 \right]$$

$$L(100) = \frac{1}{3} \left[ 2.5\text{ Billion} + 4.9\text{ Billion} + 8.1\text{ Billion} \right]$$

$$L(100) = \frac{15.5 \text{ Billion}}{3} \approx \mathbf{5,166,666,667}$$

---

### 4. Taking the Derivative (Calculus)

We want to find how the Loss ($L$) changes if we nudge the weight ($w$). Mathematically, this is the partial derivative: **$\frac{\partial L}{\partial w}$**.

Using the **Chain Rule** from calculus:
$$\frac{d}{dw}[u^2] = 2u \cdot \frac{du}{dw}$$

Let $u = (w \cdot x_i - y_i)$. 
The derivative of $u$ with respect to $w$ is simply $x_i$ (since $y_i$ is a constant, and the derivative of $w \cdot x_i$ with respect to $w$ is $x_i$).

Therefore, the derivative for any single house $i$ is:
$$\frac{\partial}{\partial w}(w \cdot x_i - y_i)^2 = 2(w \cdot x_i - y_i) \cdot x_i$$

Now, we apply this to our entire Mean Squared Error sum:

$$\frac{\partial L}{\partial w} = \frac{2}{N} \sum_{i=1}^{N} (w \cdot x_i - y_i) \cdot x_i$$

---

### 5. Plugging the Numbers into the Derivative ($w = 100$)

Let’s calculate this value for our three houses when our weight $w = 100$:

$$\frac{\partial L}{\partial w} = \frac{2}{3} \left[ \underbrace{(100 \cdot 1000 - 150,000) \cdot 1000}_{\text{House 1}} + \underbrace{(100 \cdot 1500 - 220,000) \cdot 1500}_{\text{House 2}} + \underbrace{(100 \cdot 2000 - 290,000) \cdot 2000}_{\text{House 3}} \right]$$

$$\frac{\partial L}{\partial w} = \frac{2}{3} \left[ (-50,000) \cdot 1000 + (-70,000) \cdot 1500 + (-90,000) \cdot 2000 \right]$$

$$\frac{\partial L}{\partial w} = \frac{2}{3} \left[ -50,000,000 - 105,000,000 - 180,000,000 \right]$$

$$\frac{\partial L}{\partial w} = \frac{2}{3} \left[ -335,000,000 \right]$$

$$\frac{\partial L}{\partial w} \approx \mathbf{-223,333,333.3}$$

---

### 6. Decoding What This Result Means

We calculated:
$$\frac{\partial L}{\partial w} \approx -223.3 \text{ Million}$$

**It means -223.3 is not a change in unit, it is a change in rate of error with respect to weight 100.**<br>
**so it simply means, with each 1 unit of weight the error will be off by -223.3 (reduced at rate 223.3)**

1. **It is NEGATIVE:** Because the slope is negative, calculus tells us that **increasing** $w$ will **decrease** the total Loss.
2. **It is HUGE:** The value is extremely large because our house prices are in the hundreds of thousands and sizes are in the thousands. 

#### The Weight Update Step (Gradient Descent)
To update the weight, we use the Gradient Descent rule:
$$w_{\text{new}} = w_{\text{old}} - \eta \cdot \frac{\partial L}{\partial w}$$

*(Where $\eta$ is our **learning rate**—a tiny scale factor to make sure we don't overshoot. Let's set a very small learning rate like $\eta = 0.00000018$)*:

$$w_{\text{new}} = 100 - (0.00000018 \cdot -223,333,333.3)$$

$$w_{\text{new}} = 100 - (-40.2)$$

$$w_{\text{new}} = \mathbf{140.2}$$

By calculating the derivative over the dataset, the math told us *exactly* how much to adjust our weight knob: from **100** up to **140.2**, which is almost the perfect value to minimize our error!
