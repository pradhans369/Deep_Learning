```python
import keras
import numpy as np
from keras import layers, Sequential
import keras_hub
```

### Importing Dataset from Keras Directly

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
```

```python
print(f"NUMBER OF IMAGES IN TRAINING DATA : {len(x_train)}")
print(f"NUMBER OF IMAGES IN TEST DATA : {len(x_test)}")
```

### Importing Pre-Trained Model

```python
model = keras_hub.models.ImageClassifier.from_preset("resnet_vd_50_ssld_v2_imagenet")
model.trainable = False
```
