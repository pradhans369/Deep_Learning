# Perceptron for Regression

While the classic Perceptron is designed for binary classification (using a step function as the activation), we can adapt its core structure for **regression problems** by modifying the activation function. Instead of a step function, we use a **linear activation function** (often called a Linear Neuron or ADALINE without the quantization step). 

## The Formula

For a given input vector $\mathbf{x} = (x_1, x_2, ..., x_n)$ and weights $\mathbf{w} = (w_1, w_2, ..., w_n)$ with bias $b$:

1. **Net Input:**
   $$ z = \sum_{i=1}^{n} w_i x_i + b  $$

2. **Activation Function (Linear):**
   For regression, the output $\hat{y}$ is simply the net input itself.
   $$ \hat{y} = f(z) = z $$

3. **Loss Function (Mean Squared Error):**
   To train this, we typically minimize the Mean Squared Error (MSE) between the predicted value $\hat{y}$ and the true continuous value $y$.
   $$ L = \frac{1}{2N} \sum (\hat{y} - y)^2 $$

4. **Weight Update Rule (Gradient Descent):**
   $$ w_i \leftarrow w_i - \eta \frac{\partial L}{\partial w_i} = w_i + \eta (y - \hat{y}) x_i $$
   $$ b \leftarrow b + \eta (y - \hat{y}) $$
   *(where $\eta$ is the learning rate)*

---

## Example Code (Python)

Here is a simple implementation of a single linear neuron (regression perceptron) using standard Python.

```python
import numpy as np
from sklearn.linear_model import SGDRegressor, LinearRegression

# 1. Generate dummy data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1).flatten()

# --- OPTION A: The "Gradient Descent" way (Matches our manual code) ---
# This is what you would use for very large datasets
sgd_model = SGDRegressor(max_iter=1000, tol=1e-3, eta0=0.01)
sgd_model.fit(X, y)

print("--- SGDRegressor (Gradient Descent) ---")
print(f"Weight: {sgd_model.coef_[0]:.4f}, Bias: {sgd_model.intercept_[0]:.4f}")

# --- OPTION B: The Standard way (Most Common) ---
# This is faster and more accurate for most normal-sized problems
lin_reg = LinearRegression()
lin_reg.fit(X, y)

print("\n--- Standard LinearRegression ---")
print(f"Weight: {lin_reg.coef_[0]:.4f}, Bias: {lin_reg.intercept_[0]:.4f}")

# 2. Make a prediction
X_new = np.array([[1.5]])
prediction = lin_reg.predict(X_new)
print(f"\nPrediction for X=1.5: {prediction[0]:.4f}")



## Using a Library (Scikit-Learn)

You might be wondering: *"Why do we have to build this manually? Isn't there a library for this?"*

The term **Perceptron** is historically reserved strictly for **classification** (specifically using a step function). When you remove the step function and use a linear activation to predict continuous values, you are essentially just doing **Linear Regression trained with Gradient Descent**.

Because of this terminology, libraries like `scikit-learn` do not have a `RegressionPerceptron`. Instead, you would use:
- `sklearn.linear_model.LinearRegression` (Uses the Ordinary Least Squares exact mathematical solution)
- `sklearn.linear_model.SGDRegressor` (Uses Gradient Descent to find the weights, which is **exactly what the manual code above does**)

Here is how you do the exact same thing using `scikit-learn`'s `SGDRegressor`:

```python
import numpy as np
from sklearn.linear_model import SGDRegressor

# 1. Generate the same linear dummy data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
y = y.flatten()

# 2. Train the model using Stochastic Gradient Descent (SGD)
# This is the library equivalent of our manual "Regression Perceptron"
# Using squared_error loss makes it identical to our manual math
model = SGDRegressor(loss='squared_error', learning_rate='constant', eta0=0.01, max_iter=1000)
model.fit(X, y)

# 3. View the learned weights and bias
print(f"Learned Weight: {model.coef_[0]:.4f} (True: 3)")
print(f"Learned Bias: {model.intercept_[0]:.4f} (True: 4)")

# 4. Make a prediction
X_test = np.array([[1.5]])
print(f"Prediction for X=1.5: {model.predict(X_test)[0]:.4f}")
```
