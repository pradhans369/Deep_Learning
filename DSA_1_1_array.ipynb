# To make dynamic arrays, python do not have such feature,

Hence to use it we have to write codes in `C`, and to write code in c, we have to import `ctypes`

```python
import ctypes

class myarr:
    def __init__(self):
        self.num = 0       # Number of elements currently in the array
        self.size = 1      # Current capacity of the array
        self.arr = self.make_array(self.size)

    # Assigns the capacity and returns a raw array
    def make_array(self, size):
        temp = (size * ctypes.py_object)()
        return temp

    # Returning element at a certain index (using get method)
    def get(self, index):
        if not 0 <= index < self.num:
            raise IndexError('INDEX OUT OF BOUNDS')
        return self.arr[index]

    # OPTIONAL BONUS: Allows using brackets like arr[index] directly!
    def __getitem__(self, index):
        return self.get(index)

    # Returning size of the array
    def __len__(self):
        return self.num

    # Fixed to single underscore to match the call in append()
    def _resize(self, new_size):
        new_arr = self.make_array(new_size)

        for i in range(self.num):
            new_arr[i] = self.arr[i]

        self.arr = new_arr
        self.size = new_size

    # Appending number (renamed parameter to 'item' to avoid confusion with self.num)
    def append(self, item):
        if self.num == self.size:
            self._resize(2 * self.size)        

        self.arr[self.num] = item
        self.num += 1
```

```python
arr = myarr()
```

```python
arr.append(1)
arr.append(10)
arr.append(100)
```
