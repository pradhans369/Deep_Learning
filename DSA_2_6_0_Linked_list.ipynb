```python
# making a linked list

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None
        self.n = 0          # total number of nodes in linked list

    def __len__(self):
        return self.n

    # Prints the linked list items
    def __str__(self):
        curr = self.head
        result = ""
        while curr is not None:
            result += f"{curr.data} -> "
            curr = curr.next
        
        result += "None"
        return result

    # ==================== INSERTION OPERATIONS ====================

    # inserting item to the head
    def insert_head(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node
        self.n += 1
    
    # appending value at the end of the linked list
    def append(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
        else:
            curr = self.head
            while curr.next is not None:
                curr = curr.next
            curr.next = new_node
        
        self.n += 1


```

```python
n1 = Node(1)
n2 = Node(2)
n3 = Node(3)

n1.next = n2
n2.next = n3

```

```python
print(n1.next.data)
print(n2.next.data)
print(n3.next)          # will show None
```

```python
L = LinkedList()

L.insert_head(1)
L.insert_head(2)
L.insert_head(3)
L.insert_head(4)
```
