```python
class Node:
    """A Node represents a single element in the Linked List."""
    def __init__(self, data):
        self.data = data
        self.next = None  # Pointer to the next node


class LinkedList:
    """Singly Linked List containing CRUD operations and traversal."""
    def __init__(self):
        self.head = None  # Pointer to the first node
        self.n = 0        # Count of nodes in the list
    
    def __len__(self):
        """Returns the size of the list. O(1) time."""
        return self.n

    def __str__(self):
        """Visualizes the list as: 1 -> 2 -> 3 -> None. O(N) time."""
        curr = self.head
        result = ""
        while curr is not None:
            result += f"{curr.data} -> "
            curr = curr.next
        result += "None"
        return result

    # ==================== INSERTION OPERATIONS ====================

    def insert_head(self, value):
        """Inserts a new node at the very beginning. O(1) time."""
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node
        self.n += 1

    def append(self, value):
        """Inserts a new node at the end of the list. O(N) time."""
        new_node = Node(value)
        
        # If list is empty, new node becomes the head
        if self.head is None:
            self.head = new_node
        else:
            # Traverse to the last node
            curr = self.head
            while curr.next is not None:
                curr = curr.next
            curr.next = new_node            
        self.n += 1

    def insert_after(self, target_value, value):
        """Inserts a new node after a specific value. O(N) time."""
        curr = self.head
        while curr is not None:
            if curr.data == target_value:
                break
            curr = curr.next
            
        if curr is None:
            raise ValueError(f"Value '{target_value}' not found in the list.")
            
        new_node = Node(value)
        new_node.next = curr.next
        curr.next = new_node
        self.n += 1

    def insert_at(self, index, value):
        """Inserts a new node at a specific index. O(N) time."""
        if not 0 <= index <= self.n:
            raise IndexError("Index out of range.")
            
        if index == 0:
            self.insert_head(value)
            return

        new_node = Node(value)
        curr = self.head
        for _ in range(index - 1):
            curr = curr.next
            
        new_node.next = curr.next
        curr.next = new_node
        self.n += 1

    # ==================== DELETION OPERATIONS ====================

    def delete_head(self):
        """Removes the first node. O(1) time."""
        if self.head is None:
            raise IndexError("Delete from empty list.")
            
        self.head = self.head.next
        self.n -= 1

    def pop(self):
        """Removes the last node. O(N) time."""
        if self.head is None:
            raise IndexError("Pop from empty list.")
            
        # If there is only one element
        if self.head.next is None:
            self.head = None
        else:
            # Traverse to the second-to-last node
            curr = self.head
            while curr.next.next is not None:
                curr = curr.next
            curr.next = None
            
        self.n -= 1

    def delete_value(self, value):
        """Removes a node with a specific value. O(N) time."""
        if self.head is None:
            raise ValueError("Delete from empty list.")
            
        # Case 1: Target value is at the head
        if self.head.data == value:
            self.head = self.head.next
            self.n -= 1
            return

        # Case 2: Target is in the middle or end
        curr = self.head
        while curr.next is not None:
            if curr.next.data == value:
                curr.next = curr.next.next
                self.n -= 1
                return
            curr = curr.next
            
        raise ValueError(f"Value '{value}' not found in the list.")

    # ==================== UTILITY OPERATIONS ====================

    def search(self, value):
        """Searches for a value. Returns its index if found, else -1. O(N) time."""
        curr = self.head
        index = 0
        while curr is not None:
            if curr.data == value:
                return index
            curr = curr.next
            index += 1
        return -1

    def reverse(self):
        """Reverses the linked list in-place (no extra memory). O(N) time."""
        prev = None
        curr = self.head
        
        while curr is not None:
            next_node = curr.next  # Save the rest of the list
            curr.next = prev       # Reverse the link
            prev = curr            # Move 'prev' one step forward
            curr = next_node       # Move 'curr' one step forward
            
        self.head = prev           # Update the head to the new front

# ==================== VERIFICATION & TESTING ====================
```

```python
ll = LinkedList()

# Testing Insertions
ll.insert_head(10)
ll.insert_head(5)
ll.append(20)
ll.append(30)
ll.insert_after(20, 25)
ll.insert_at(2, 15)

print("Initial list:", ll)       # Expected: 5 -> 10 -> 15 -> 20 -> 25 -> 30 -> None
print("List size:", len(ll))      # Expected: 6

# Testing Search
print("Search for 25 (Index):", ll.search(25))  # Expected: 4

# Testing Deletions
ll.delete_head()
print("After delete head:", ll)  # Expected: 10 -> 15 -> 20 -> 25 -> 30 -> None

ll.pop()
print("After pop (tail):", ll)   # Expected: 10 -> 15 -> 20 -> 25 -> None

ll.delete_value(20)
print("After delete '20':", ll) # Expected: 10 -> 15 -> 25 -> None

# Testing Reverse
ll.reverse()
print("Reversed list:", ll)      # Expected: 25 -> 15 -> 10 -> None
```
