```python
clean_stories = [story.strip() for story in train[:] if story.strip() != ""]
```

```python
def clean_text(text):
    text = text.lower()
    text = text.translate(str.maketrans("", "",string.punctuation))
    return text.split()
```

str.maketrans(x, y, z)
This function creates a translation map (a dictionary of character mappings). It takes up to three arguments:

- x: Characters you want to replace.
- y: Characters you want to replace them with.
- z: Characters you want to delete entirely.

We pass empty strings "" for x and y because we do not want to replace any characters. We pass string.punctuation as the third parameter z because we want to delete all punctuations.

```python
tokenization_stories = [clean_text(story) for story in clean_stories]
```

```python
word_counts = {}
for story in tokenization_stories:
    for word in story:
        word_counts[word] = word_counts.get(word, 0) + 1
```

```python
list(word_counts.items())[:10]
```

```python
sorted_words = sorted(word_counts.items(), key=lambda x:x[1], reverse=True)
```

Based on the number of the word, the entire list of the words will be sorted.

```python
vocab_size = 10000
top_words = [word for word, count in sorted_words[: vocab_size - 1]]
```

```python
top_words[:10]
```

```python
word2idx = {word: idx + 1 for idx, word in enumerate(top_words)}
word2idx["<UNK>"] = 0
idx2word = ["<UNK>"] + top_words
```

```python
seq_len = 10

# 1. Define the sequences using your existing clean_text function
sequences = [[word2idx.get(w, 0) for w in clean_text(story)] for story in train[:2000]]

# 2. Generate the sliding window input sequences
input_sequences = [seq[i:i + seq_len + 1] for seq in sequences for i in range(len(seq) - seq_len)]

# 3. Create x and y
# x is the first seq_len tokens, y is the target (the next token)
input_data = np.array(input_sequences)
x, y = input_data[:, :-1], input_data[:, -1]
```

```python
print(x)
print()
print()
print(y)
```

```python
import keras
from keras import layers, Sequential
from keras.layers import LSTM, Dense, Embedding
```

```python
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=50, input_shape=(sequence_length,)),
    LSTM(128),
    Dense(vocab_size, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(x, y, epochs=10, batch_size=64)
```

```python

def generate_text(seed_text, next_words=10):
    for _ in range(next_words):
        cleaned_seed = clean_text(seed_text)
        seed_seq = [word2idx.get(word, 0) for word in cleaned_seed]
        if len(seed_seq) < sequence_length:
            seed_seq = [0] * (sequence_length - len(seed_seq)) + seed_seq
        else:
            seed_seq = seed_seq[-sequence_length:]
        
        input_data = np.array([seed_seq])
        prediction = model.predict(input_data, verbose=0)
        predicted_idx = np.argmax(prediction, axis=-1)[0]
        predicted_word = idx2word[predicted_idx]
        seed_text += " " + predicted_word
    return seed_text

print(generate_text("once upon a time there was a little"))
```
