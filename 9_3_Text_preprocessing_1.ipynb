### Using `text_to_word_sequence`
- Used to convert text into words

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import text_to_word_sequence
```

```python
temp = """Hello!!! i am the master of all the DISGuISE...., Behold the true self of Ashura!@^% ^&Q(@qwre 27361%^&92 ___  hel|lo. 
I hope you all are doing fine here with all the wealth you have.
"""

ans = text_to_word_sequence(temp)
print(ans)

```

It is good at text preprocessing but, there is one disadvantage of it, which is, "hel|lo" here instead of just removing the symbol, it just removed the symbols and separated both the sides into different words. Which is NOT RECOMMENDED while text preprocessing.


```python
# To regex to solve the above problem

import re
ans = re.sub(r'[^\w\s]','', temp).lower()
print(ans)
```

See the difference clearly

```python
# Removing the lines and putting every word into a list
ans = re.sub(r'[^\w\s]','', temp).lower().replace('\n','')
print(ans, '\n')

ans = text_to_word_sequence(ans)
print(ans)
```

### Applying text preprocessing on `para`

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import text_to_word_sequence

import re
```

```python
def text_preprocessing(text):
    text = re.sub(r'[^\w\s]','',text).lower().replace('\n','')
    return text_to_word_sequence(text)


preprocessed_para = text_preprocessing(para)
```

```python
print(preprocessed_para)
```

### Setting frequence to each word

```python
freq = {}

for word in preprocessed_para:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1
```

```python
print(type(freq))
print(freq)
```

```python
print(list(freq)[:10])
print(list(freq.items())[:10])

print(max(freq.values()))
print(min(freq.values()))
```

```python
# Printing the words which are with more than 5 frequency
print([(keys, values) for keys, values in freq.items() if values > 3])
```

```python
for i,j in freq.items():
    if j > 3:
        print((i, j))
```

```python
# Putting top 20 words in an obj
top_words = sorted(freq.items(), key=lambda x : x[1], reverse=True)[:20]
print(top_words)
```

```python
# Printing top 20 words from the list of touples
print([
    touple[0] for touple in top_words
])
```
