# Regularization


Before adding regularization, standard Linear Regression simply tries to minimize the **Mean Squared Error (MSE)**. 
Here is the base formula for standard MSE Loss:

$$ \text{Loss} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

To prevent the model from overfitting, Regularization adds a "Penalty Term" to the end of this Loss function. The model is now forced to minimize the error *and* keep the weights small at the same time.


### 1. Ridge Regression (L2 Regularization)
Ridge adds the **squared magnitude** of the weights to the loss function. 
*   **Effect:** It shrinks all weights to be very small, but rarely forces them to be exactly 0. 

$$ \text{Loss}_{Ridge} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{p} w_j^2 $$

### 2. Lasso Regression (L1 Regularization)
Lasso adds the **absolute value** of the weights to the loss function.
*   **Effect:** It acts as an automatic feature selector. It will literally force the weights of useless features to become exactly 0, removing them from the equation entirely.

$$ \text{Loss}_{Lasso} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{p} |w_j| $$


**What do the symbols mean?**
*   $y_i$ = The actual, true answer
*   $\hat{y}_i$ = The model's prediction
*   $\lambda$ (Lambda) = The tuning dial. If $\lambda = 0$, there is no penalty (it's just normal regression). If $\lambda$ is huge, the model heavily prioritizes shrinking the weights.
*   $w_j$ = The weight (coefficient) for a specific feature. Here the value for $w$ will be the same as the value of $w$ in $Loss$ function for $\hat{y} = wx + b$
