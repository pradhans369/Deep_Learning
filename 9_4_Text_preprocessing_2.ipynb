### Storig all the row data from the dataset into a single obj

```python
import re
para = [[sentence.lower()] for sentence in data['text']]
para = [[re.sub(r'[^\w\s]','',chars)] for sentence in para for chars in sentence]
```

### Text Preprocessing : Converting the entire sentences into list of words

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import text_to_word_sequence

import re
```

```python
for i in para[:5]:
    print(i[0])
```

```python
para_preprocessed = [text_to_word_sequence(sentence[0]) for sentence in para]
```

```python
para_preprocessed[:2]
```

### Checking for total number of words

```python
# words and their frequency
freq = {}

for sentence in para_preprocessed:
    for word in sentence:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1

```

```python
# printing total number of words
print(len(freq))

# printing top 20 high frequency words
print(sorted(freq.items(), key=lambda x : x[1], reverse=True)[:20])
```

### Applying vectorization

### Approach 1 : Using `Tokenizer`

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer(num_words=5000,oov_token="<UNK>")

tokenizer.fit_on_texts(para_preprocessed)
para_vectors = tokenizer.texts_to_sequences(para_preprocessed)
```

```python
para_vectors[:2]
```

### Approach 2 : Using `TextVectorization`

```python
from tensorflow.keras.layers import TextVectorization
```

Since TextVectorization only expects stirng, so we had to send `para` instead of `para_preprocessed` for `TextVectorization`

```python
vectorizer = TextVectorization(max_tokens=5000)
vectorizer.adapt(para)
para_vector = vectorizer(para).numpy()
```
