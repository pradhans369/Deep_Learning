### Text Preprocessing

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import text_to_word_sequence

import re
```

```python
def preprocessing(text):
    text = re.sub(r'[^\w\s]','',text).lower()
    return text_to_word_sequence(text)


single_para_preprocessed = preprocessing(single_para)
```

### Putting the frequency to each word

```python
freq = {}

for word in single_para_preprocessed:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

print(len(freq))
```

### Putting the freq in descending order

```python
top_words = sorted(freq.items(), key=lambda x : x[1], reverse=True)
print(top_words[:15])
```
